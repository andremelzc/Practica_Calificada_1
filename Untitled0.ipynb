Para hallar la pendiente usamos esta fórmula:


image.png

```python
#Definimos en la siguiente función:
def pendiente(n,xy,x,y,powx):
  m = (n*xy-x*y)/(n*powx-x*x)
  return m
```

Para hallar el intercepto usamos esta fórmula:

image.png

```python
#Definimos en la siguiente función:
def intercepto(powx,y,x,xy,n):
  b = (powx*y-x*xy)/(n*powx-x*x)
  return b
```

**a) Ley de Boyle**

```python
#Obteniendo los datos de la gráfica "V vs P"
df = pd.read_csv("data.csv")
df
```

```python
#Cambiando los datos a "V vs 1/P"
df["y"] = 1/df["y"]
df
```

```python
#Graficando los datos "V vs 1/P"
plt.scatter(df["x"],df["y"])
```

```python
sumx1 = df["x"].sum()
sumxpow1 = (df["x"]*df["x"]).sum()
sumy1 = df["y"].sum()
sumxpy1 = (df["x"]*df["y"]).sum()
```

```python
pendiente1 = pendiente(298,sumxpy1,sumx1,sumy1,sumxpow1)
pendiente1
```

```python
intercepto1 = round(intercepto(sumxpow1,sumy1,sumx1,sumxpy1,298),4)
intercepto1
```

```python
print("Regresión lineal:",pendiente1,"x","+",intercepto1)
```

## 2. Extraer datos químicos de la web y hacer una exploración gráfica

```python
# Emisiones de CO2 en el Perú, desde el 1970 hasta el 2021
url = "https://datosmacro.expansion.com/energia-y-medio-ambiente/emisiones-co2/peru#:~:text=En%20Per%C3%BA%20se%20incrementan%20las%20emisiones%20de%20CO2&text=En%20la%20tabla%20vemos%20que,1%2C64%20toneladas%20por%20habitante."
tablas = pd.read_html(url)
tablas[0]
```

```python
# Adecuando el tamaño, la rotación y poniendo títulos a los ejes
# Gráfica de barras
plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.xlabel("Año")
plt.ylabel("Emisiones de CO2 en toneladas")
plt.title("Emisiones de CO2 en el Perú")
plt.bar(tablas[0]['Fecha'],tablas[0]['CO2 Totales Mt'])
```

```python
# Diagrama de cajas y bigotes
# Sirve para visualizar los cuartiles
plt.boxplot(tablas[0]['CO2 Totales Mt'])
```

```python
# Diagrama de violin
plt.violinplot(tablas[0]['CO2 Totales Mt'])
```

```python
# Hallando las medidas de tendencia central
medidas = round(tablas[0]["CO2 Totales Mt"].describe(),2)
# Dándole forma "bonita" y "más entendible" a la impresión
print("Número de datos:",medidas[0])
print("Media:",medidas[1])
print("Desviación estándar:",medidas[2])
print("Dato mínimo:",medidas[3])
print("Primer cuartil",medidas[4])
print("Mediana:",medidas[5])
print("Tercer cuartil:",medidas[6])
print("Dato máximo:",medidas[7])
```
