### Pregunta 1: DataFrames y listas en python (4 puntos)
Usando PeriodicTable, crea una tabla con 20 elementos químicos y en las columnas considera: nombre, simbolo, masa_atomica, numero_atomico y valencias.

```python
!pip install rdkit
!pip install pandas
```

```python
from rdkit import Chem
PeriodicTable = Chem.GetPeriodicTable()
import pandas as pd
```

```python
simbolo_elemento = []
for i in range (1,21):
  simbolo_elemento.append(PeriodicTable.GetElementSymbol(i))

nombre_elemento = [PeriodicTable.GetElementName(i) for i in range(1,21)]
masa_atomica = [PeriodicTable.GetAtomicWeight(i) for i in simbolo_elemento]
num_atomico = [PeriodicTable.GetAtomicNumber(i) for i in simbolo_elemento]
valencia = [PeriodicTable.GetValenceList(i) for i in simbolo_elemento]

info = {"Num Atómico":num_atomico,"Nombre":nombre_elemento,"Símbolo":simbolo_elemento,"Masa Atómica":masa_atomica,"Valencia":valencia}
tabla = pd.DataFrame(info)
tabla
```
